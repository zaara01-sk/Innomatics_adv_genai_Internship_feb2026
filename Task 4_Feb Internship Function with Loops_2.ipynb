## Problem 1: Mobile Recharge Validation System

```python
def validate_recharge(amount, plan):
    valid_plans = [199, 299, 399, 599]
    if amount >= 50 and plan in valid_plans:
        return True
    return False

while True:
    try:
        amt = int(input("Enter recharge amount (Min ₹50): "))
        pln = int(input("Enter plan (199, 299, 399, 599): "))
        
        if validate_recharge(amt, pln):
            print("Recharge Successful!")
            break 
        else:
            print("Invalid Amount or Plan. Please retry.")
    except ValueError:
        print("Please enter valid numbers.")
```

## Problem 2: Inventory Reorder Alert System

```python
def check_inventory(stock_dict):
    for product, quantity in stock_dict.items():
        if quantity < 15:
            print(f"Alert: {product} (Stock: {quantity}) - Reorder Alert")
        else:
            print(f"Status: {product} (Stock: {quantity}) - Stock OK")

# Sample Data
inventory = {"Laptop": 20, "Mouse": 10, "Keyboard": 15, "Monitor": 5}
check_inventory(inventory)
```

## Problem 3: Student Result Processing System

```python
def process_student_result(marks_list):
    total = 0
    for mark in marks_list:
        total += mark
    
    average = total / len(marks_list)
    return "Pass" if average >= 50 else "Fail."

# Example usage
marks = [65, 70, 45, 80]
print(f"Final Result: {process_student_result(marks)}")
```

## Problem 4: Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, is_peak):
    base_fare = 50
    per_km_rate = 12
    total = base_fare + (distance * per_km_rate)
    
    if is_peak.lower() == 'yes':
        total *= 1.25  # 25% surge
    return total

while True:
    try:
        dist = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ")
        print(f"Estimated Fare: ₹{calculate_fare(dist, peak):.2f}")
        
        if input("Calculate another? (y/n): ").lower() != 'y':
            break
    except ValueError:
        print("Invalid distance.")
```

## Problem 5: Employee Attendance Eligibility Checker

```python
def check_eligibility(attendance_list):
    present_days = 0
    for status in attendance_list:
        if status.upper() == "P":
            present_days += 1
            
    percentage = (present_days / len(attendance_list)) * 100
    return "Eligible" if percentage >= 75 else "Not Eligible"

# Example usage
attendance = ["P", "P", "A", "P", "P", "P", "A"]
print(f"Attendance Status: {check_eligibility(attendance)}")
```

## Problem 6: Password Strength Checker

```python
def check_password_strength(password):
    if len(password) < 8:
        return "Weak: Too short (min 8 chars)"
    
    has_digit = False
    has_special = False
    special_chars = "@#$"
    
    for char in password:
        if char.isdigit():
            has_digit = True
        elif char in special_chars:
            has_special = True
            
    if has_digit and has_special:
        return "Strong Password"
    return "Weak: Missing digit or special character (@#$)"

# Example test
print(check_password_strength(input("Enter password: ")))
```
