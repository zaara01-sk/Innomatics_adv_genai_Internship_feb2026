## Problem 1: Social Media – Post Engagement Analyzer 

```python
def post_engagement_analyzer(likes_list):
    total_likes = sum(likes_list) 
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    return total_likes, status

likes = [500, 300, 250]
total, status = post_engagement_analyzer(likes)
print(f"Total Likes: {total}\nPost Status: {status}")
```

## Problem 2: Healthcare – Medicine Stock Alert System 

```python
def medicine_stock_alert(stock):
    if stock < 10:
        return "Low Stock Alert."
    else:
        return "Stock Sufficient."

print(f"Status: {medicine_stock_alert(6)}")
```

## Problem 3: Agriculture – Rainfall Adequacy Checker 

```python
def rainfall_adequacy_checker(rainfall_data, required_level):
    average_rainfall = sum(rainfall_data) / len(rainfall_data) 
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    return average_rainfall, status

avg, status = rainfall_adequacy_checker([80, 65, 71], 70)
print(f"Average Rainfall: {avg:.1f}\nRainfall Status: {status}")
```

## Problem 4: Social Media – Duplicate Account Detection 

```python
def detect_duplicates(usernames):
    if len(usernames) != len(set(usernames)):
        return "Yes"
    return "No"

print(f"Duplicate Accounts Found: {detect_duplicates(['user1', 'user2', 'user1'])}")
```

## Problem 5: Healthcare – Appointment Eligibility Checker 

```python
def check_appointment_eligibility(age):
    if age >= 18:
        return "Eligible"
    else:
        return "Not Eligible."

print(f"Eligibility Status: {check_appointment_eligibility(21)}")
```

## Problem 6: Agriculture – Premium Crop Price Filter 

```python
def filter_premium_crops(prices):
    premium_crops = [price for price in prices if price > 2000]
    return premium_crops

print(f"Premium Crops: {filter_premium_crops([1500, 2500, 1800, 3200])}")
```

## Problem 7: System Monitoring – Application Health Checker 

```python
def application_health_checker(errors):
    if errors == 0:
        return "Healthy"
    elif errors <= 5:
        return "Minor Issues."
    else:
        return "Critical Issues."

print(f"System Status: {application_health_checker(7)}")
```

## Problem 8: Banking – Daily Transaction Limit Checker 

```python
def daily_limit_checker(amount):
    limit = 50000
    if amount <= limit:
        return "Approved"
    else:
        return "Rejected"

print(f"Transaction Status: {daily_limit_checker(60000)}")
```

## Problem 9: E-Learning – Student Attendance Eligibility System 

```python
def check_exam_eligibility(attendance_list):
    percentage = (sum(attendance_list) / len(attendance_list)) * 100
    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"
    return percentage, eligibility

perc, status = check_exam_eligibility([1, 1, 1, 0, 1])
print(f"Attendance Percentage: {perc:.1f}\nExam Eligibility: {status}")
```

## Problem 10: Smart Electricity Bill Analyzer 

```python
def electricity_bill_analyzer(units):
    # Slab calculations 
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7
    
    # Usage Classification 
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
        
    return bill, status

bill, status = electricity_bill_analyzer(250)
print(f"Total Bill: ₹{bill}\nUsage Level: {status}")
```
