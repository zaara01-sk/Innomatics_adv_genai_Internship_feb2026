## Problem Statement 1:Smart Parking Lot Management System

```python
def smart_parking(capacity, logs):
    current = 0
    peak = 0

    for log in logs:
        if log == "IN":
            current += 1
        elif log == "OUT":
            current -= 1

        if current > peak:
            peak = current

    print(f"Currently Parked Vehicles: {current}")
    print(f"Peak Parking Usage: {peak}")

    if current > capacity:
        print("Parking Status: Full - ALERT! Capacity Exceeded")
    else:
        print("Parking Status: Available")

# Sample Input
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
smart_parking(capacity, logs)
```

## Problem Statement 2:Online Food Delivery Time Estimator 

```python
def estimate_delivery_time(distance, traffic, weather):
    base_time = (distance / 30) * 60
    if traffic == "Low":
        traffic_delay = 0
    elif traffic == "Medium":
        traffic_delay = 10
    elif traffic == "High":
        traffic_delay = 20
    else:
        traffic_delay = 0

    # Weather delay adjustments
    if weather == "Clear":
        weather_delay = 0
    elif weather == "Cloudy":
        weather_delay = 5
    elif weather == "Rainy":
        weather_delay = 15
    else:
        weather_delay = 0

    total_time = round(base_time + traffic_delay + weather_delay)
    print(f"Estimated Delivery Time: {total_time} minutes")

# Sample Input
estimate_delivery_time(distance=8, traffic="High", weather="Rainy")
```

## Problem Statement 3: Movie Theatre Seat Occupancy Analyzer

```python
def theatre_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy_percent = (booked_count / total_seats) * 100
    print(f"Occupancy: {occupancy_percent:.0f}%")

    if occupancy_percent == 100:
        print("Show Status: Housefull")
        print("Suggestion: Consider opening additional shows!")
    elif occupancy_percent >= 75:
        print("Show Status: Almost Full")
        print("Suggestion: Consider opening additional shows!")
    else:
        print("Show Status: Available")

# Sample Input
total_seats = 200
booked_seats = [1] * 150  # 150 booked seats
theatre_occupancy(total_seats, booked_seats)
```

## Problem Statement 4: Cloud Server Load Classification System

```python
def classify_server_load(cpu_readings):
    avg_cpu = sum(cpu_readings) / len(cpu_readings)
    print(f"Average CPU Load: {avg_cpu:.0f}%")

    if avg_cpu < 50:
        print("Server Status: Normal")
    elif 50 <= avg_cpu <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")

# Sample Input
cpu_readings = [45, 60, 70, 85, 90]
classify_server_load(cpu_readings)
```

## Problem Statement 5:Smart Classroom Resource Usage Monitor

```python
def monitor_classroom_resources(usage, threshold=8):
    overused = []
    for resource, hours in usage.items():
        if hours >= threshold:
            overused.append(resource)

    if overused:
        print(f"Overused Resources: {', '.join(overused)}")
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

# Sample Input
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
monitor_classroom_resources(usage)
```

## Problem Statement 6: Online Event Registration Capacity Controller 

```python
def manage_event_registration(capacity, registrations):
    
    if registrations <= capacity:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"

    print(f"Confirmed Registrations: {confirmed}")
    print(f"Waitlisted Users: {waitlisted}")
    print(f"Registration Status: {status}")

# Sample Input
manage_event_registration(capacity=100, registrations=105)
```
