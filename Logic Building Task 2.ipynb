# Problem 1: Unique Words in a sentence

```python
# sentence =  input("Enter sentence: ").strip() 
sentence = "Python is easy and Python is powerful" # hardcoded sentence because the program was crashing as I entered "o".
words = sentence.lower().split()
unique_words = set(words) #removes duplicates
print(f"Unique words count: {len(unique_words)}")
print(f"Unique words: {unique_words}")
```

# Problem 2: Highest salary from Employee Data

```python
employees = { "Ravi": 75000, "Anita": 68000, "Kiran": 72000}
highest_employee = max(employees, key = employees.get)
highest_salary = employees[highest_employee]
print(f"Highest Salary: {highest_employee} - {highest_salary}") 
```

# Problem 3: Find Maximum and Minimum Values

```python
numbers = [45,22,89,10,66]
# initialize min and max value as first element
max_val = numbers[0]
min_val = numbers[0]
# finding min and max via loop
for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num
print(f"List: {numbers}")
print(f"Max: {max_val}")
print(f"Min: {min_val}")
```

# Problem 4: Count Products Above a Price Threshold

```python
prices = [450,1200,899,1500,300]
threshold = 1000
count = 0

for price in prices:
    if price > threshold:
        count += 1
print(f"Products above 1000: {count}")
```

# Problem 5: Calculate Attendance Percentage

```python
attendance = ["P","P","A","P","P"]
# Count "P"
present_days = attendance.count("P")
total_days = len(attendance)
# Dividing by total days for percentage
percentage = (present_days / total_days)*100
print(f"Attendance Percentage: {percentage}")
```

# Problem 7: Count Character Frequency

```python
#text = input("Enter a string: ")
text = "ppython"
frequency = {}
# counting by incrementing
for char in text:
    frequency[char] = frequency.get(char,0) + 1
print(frequency)
```
