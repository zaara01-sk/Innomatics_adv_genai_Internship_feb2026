## Problem Statement 1: Employee Performance Bonus Eligibility

```python
employees = {"Ravi": 92, "Anita":88, "Kiran":92, "Suresh":85}
max_score  = max(employees.values())
top_performers = [name for name, score in employees.items() if score == max_score]
print(f"Top performers eligible for bonus: {','. join(top_performers)} (Score: {max_score})")
```

## Problem Statement 2: Search Query Keyword Analysis

```python
import string
query = "Buy mobile phone buy phone online"
clean_query = query.lower().translate(str.maketrans('', '', string.punctuation))
words =  clean_query.split()
word_counts = {word: words.count(word) for word in set(words)}
repeated_keywords = {word: count for word, count in word_counts.items() if count > 1}
print(repeated_keywords)
```

## Problem Statement 3: Sensor Data Validation

```python
sensor_readings = [3,4,7,8,10,12,5]
valid_readings = [(index, val) for index, val in enumerate(sensor_readings) if val%2==0]
print("Valid Sensor Readings (Hour, Value): ")
print(valid_readings)
```

## Problem Statement 4: Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
domains = [email.split('@')[1] for email in emails]
domain_counts = {domain: domains.count(domain) for domain in set(domains)}
total_users = len(emails)
for domain, count in domain_counts.items():
    percentage = (count/total_users)*100
    print(f"{domain}: {int(percentage)}%")
```

## Problem Statement 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
avg_sales = sum(sales)/len(sales)
threshold = avg_sales * 1.3 # detect spikes(more than 30% above average)
for i, sale in enumerate(sales):
    if sale>threshold:
        print(f"Day {i+1}: {sale}")
```

## Problem Statement 6: Duplicate User ID Detection

```python
user_ids = ["user1","user2","user1","user3","user1","user3"]
from collections import Counter
counts = Counter(user_ids)
for user, count in counts.items():
    if count>1:
        print(f"{user} {count} times")
```
