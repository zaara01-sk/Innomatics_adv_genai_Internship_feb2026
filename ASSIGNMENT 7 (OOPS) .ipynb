## Problem 1: Smart Light Controller

```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def control(self, action):
        self.status = action
        print(f"{self.name} is {self.status}")

# Hardcoded values — because input() was crashing when typed 'o'
light_name = "Bedroom Light"
action = "ON"

light = SmartLight(light_name)
light.control(action)
```

## Problem 2: Employee ID Card System

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name       = name          # Employee's full name
        self.emp_id     = emp_id        # Unique employee ID
        self.department = department    # Department name

    def display_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")


# Hardcoded Sample Values 
name       = "Rahul"
emp_id     = "EMP102"
department = "IT"

# Creating Object and Displaying Card
emp = Employee(name, emp_id, department)   # Pass three values to constructor
emp.display_card()                         # Call method to print the ID card
```

## Problem 3: Mobile Contact Record

```python
class Contact:
    def __init__(self, name, phone):
        self.name  = name    # Contact's name
        self.phone = phone   # Contact's phone number (stored as string)

    def save_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")


# Hardcoded Sample Values
name  = "Anita"
phone = "9876543210"

# Creating Object and Saving Contact
contact = Contact(name, phone)   # Object created with name and phone
contact.save_contact()           # Display the saved contact info
```

## Problem 4: Product Price Tag Generator

```python
class Product:
    def __init__(self, name, price):
        self.name  = name          # Product name (string)
        self.price = int(price)    # Price (typecasted)

    def print_price_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: \u20b9{self.price}")  


# Hardcoded Sample Values
name  = "Headphones"
price = 2499

# Creating Object and Printing Price Tag
product = Product(name, price)
product.print_price_tag()
```

## Problem 5: Movie Rating Display System

```python
class Movie:
    def __init__(self, name, rating):
        self.name   = name            # Movie title (string)
        self.rating = float(rating)   # Rating out of 5 (float)

    def display(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")


#Hardcoded Sample Values
name   = "Inception"
rating = 4.8

#Creating Object and Displaying Info
movie = Movie(name, rating)
movie.display()
```

## Problem 6: Delivery Address Manager

```python
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name   # Customer's name
        self.address       = address         # Delivery address

    def print_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")


# Hardcoded Sample Values
customer_name = "Suman"
address       = "Hyderabad"

# Creating Object and Printing Details
delivery = Delivery(customer_name, address)
delivery.print_details()
```
