## 2) Pass/Fail Analyzer

```python
marks = [45,78,90,33,60]
pass_count = 0
fail_count = 0
for mark in marks:
    if mark >=50:
        pass_count += 1 #student passed, hence increase pass_count
    else:
        fail_count += 1 #student failed, hence increase fail_count
#printing final results clearly
print(f"Total Pass Students: {pass_count}") 
print(f"Total Fail Students: {fail_count}")
```

## 3)Simple Data Cleaner

```python
names = ["Alice", "bob", "CHARLIE"]
#cleaning names
cleaned_names = [ name.strip().lower() for name in names]
print("Cleaned List:", cleaned_names)
```

## 4) Message Length Analyzer

```python
message = ["Hi", "Welcome to the platform", "OK"]
for msg in message:
    length = len(msg)
    print(f"Message : '{msg}'| Length : {length}")
    if length > 10: # flagging
        print("-> Flag: Message is longer than 10 characters")
```

## 5) Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
#Counting Occurence of "ERROR"
error_count = logs.count("ERROR")
print(f"Total error count: {error_count}")
```
